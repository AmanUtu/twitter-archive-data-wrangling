### Tidiness issues


### Issue #1:  On a df_archive dataframe the dog stage (i.e doggo, floofer, pupper, and puppo) must one column and all four columns shoud be droped.

#### Define
On a df_archive dataframe the dog stage (i.e doggo, floofer, pupper, and puppo) must one column and all four columns shoud be droped. And grouped to one column called dog stage.

```python
df_archive_clean.columns
```

For the purpose of creating the dog stage column let's reset our index.

```python
df_archive_clean = df_archive_clean.reset_index()
```

Using the folowing for loop let's merge the four dog stages column in to one dog stage column.

```python
# Creating dog_stage column

for i in range(0, len(df_archive_clean.index)):
    if df_archive_clean.loc[i, 'doggo'] == 'doggo':
        df_archive_clean.loc[i, 'dog_stage'] = 'doggo'
    elif df_archive_clean.loc[i, 'floofer'] == 'floofer':
        df_archive_clean.loc[i, 'dog_stage'] = 'floofer'
    elif df_archive_clean.loc[i, 'pupper'] == 'pupper':
        df_archive_clean.loc[i, 'dog_stage'] = 'pupper'
    elif df_archive_clean.loc[i, 'puppo'] == 'puppo':
        df_archive_clean.loc[i, 'dog_stage'] = 'puppo'
    else:
        df_archive_clean.loc[i, 'dog_stage'] = 'None'
```

Now let's delete the four columns.

```python
columns = ['doggo', 'floofer', 'pupper', 'puppo']
df_archive_clean.drop(columns, axis=1, inplace=True)
```

In our case this two tables shows detaile about twitter post so we should merge them on based on tweet_id.

using pandas merger function let's merge both table on tweet_id of the 2 tables.
